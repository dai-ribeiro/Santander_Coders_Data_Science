**Q08** - Faça um programa que dadas duas listas de mesmo tamanho, imprima o produto escalar entre elas.

OBS: produto escalar é a soma do resultado da multiplicação entre o número na posição i da lista1 pelo número na posição i da lista2, com i variando de 0 ao tamanho da lista.

```python
lista1 = [1, 4, 5]
lista2 = [2, 2, 3]
lista3 = []

for idx, item in enumerate(lista1):
    soma = lista1[idx] * lista2[idx]
    lista3.append(soma)
print(lista3)
```

**Q13** - Faça um programa que sorteia 10 números entre 0 e 100 e conte quantos números sorteados são maiores que 50.

**Q15** - Desafio 1 - Faça um programa que peça para o usuário digitar o nome e a idade de um aluno e o número de provas que esse aluno fez. Depois, o programa deve pedir para o usuário digitar as notas de cada prova do aluno. Ao final o programa deve imprimir uma lista contendo:

a. Nome do aluno na posição 0;

b. Idade do aluno na posição 1;

c. Uma lista com todas as notas na posição 2;

d. A média do aluno na posição 3;

e. True ou False, caso a média seja maior que 5 ou não, na posição 4.

Dica: Use o que você fez nos exercícios anteriores para criar esse programa.

```python
nome = input('Nome do aluno: ')
idade = int(input('Idade do aluno: '))
provas = int(input('Quantas provas: '))
notas = []
info_aluno = []
cont = 1

while cont <= provas:
    notas.append(float(input(f'Qual a nota da prova {cont}: ')))
    cont += 1

media = sum(notas) / provas

info_aluno.append(nome)
info_aluno.append(idade)
info_aluno.append(notas)
info_aluno.append(media)
info_aluno.append(True if media>=5 else False)

print(info_aluno)
```

**Q16** - Desafio 2 - Faça um programa como o do item anterior, porém que imprima a média sem considerar a maior e menor nota do aluno (nesse caso o número de provas precisa ser obrigatoriamente maior que dois).

Dica: crie uma cópia com a lista de todas as notas antes de fazer a média.

```python
nome = input('Nome do aluno: ')
idade = int(input('Idade do aluno: '))
provas = int(input('Quantas provas: '))
notas = []
info_aluno = []
cont = 1

while provas <= 2:
    provas = int(input('Quantas provas: '))

while cont <= provas:
    notas.append(float(input(f'Qual a nota da prova {cont}: ')))
    cont += 1

copia_notas = notas.copy()
copia_notas.remove(min(copia_notas))
copia_notas.remove(max(copia_notas))

# print(notas)
# print(copia_notas)
media = sum(copia_notas) / len(copia_notas)

info_aluno.append(nome)
info_aluno.append(idade)
info_aluno.append(notas)
info_aluno.append(round(media,2))
info_aluno.append(True if media>=5 else False)

print(info_aluno)
```

**Q17** - Desafio 3 - Faça um programa que pede para o usuário digitar o CPF e verifica se ele é válido. Para isso, primeiramente o programa deve multiplicar cada um dos 9 primeiros dígitos do CPF pelos números de 10 a 2 e somar todas as respostas. O resultado deve ser multiplicado por 10 e dividido por 11. O resto dessa divisão deve ser igual ao primeiro dígito verificador (10º dígito). Em seguida, o programa deve multiplicar cada um dos 10 primeiros dígitos do CPF pelos números de 11 a 2 e repetir o procedimento anterior para verificar o segundo dígito verificador.

Exemplo:

Se o CPF for 286.255.878-87 o programa deve fazer primeiro:

x = (2*10 + 8*9 + 6*8 + 2*7 + 5*6 + 5*5 + 8*4 + 7*3 + 8*2)

Em seguida, o programa deve testar se x*10%11 == 8 (o décimo número do CPF). Se sim, o programa deve calcular:

x = (2*11 + 8*10 + 6*9 + 2*8 + 5*7 + 5*6 + 8*5 + 7*4 + 8*3 + 8*2)

e verificar se x*10%11 == 7 (o décimo primeiro número do CPF).

```python
cpf = input('Digite o número do CPF (somente números): ')
dig_cpf = []

for i in cpf:
    num = int(i)
    dig_cpf.append(num)

digito1 = dig_cpf[9]
digito2 = dig_cpf[10]

cont = 10
soma = 0
for item in dig_cpf[0:9]:
    soma += item * cont
    cont -= 1

verifica_dig1 = (soma*10%11)

cont = 11
soma = 0
for item in dig_cpf[0:10]:
    soma += item * cont
    cont -= 1 

verifica_dig2 = (soma*10%11)

if (digito1 == verifica_dig1) and (digito2 == verifica_dig2):
    print('CPF válido!')
else:
    print('CPF inválido!')
```
