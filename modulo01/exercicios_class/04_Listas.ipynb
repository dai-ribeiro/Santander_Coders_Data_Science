**Q05** - Agora usando o método max() faça um programa que imprima os três maiores números de uma lista.

Dica: Use o método próprio de listas .remove().

```python
lista = [4, 5, 6, 5, -6, 56, 7, -10, 78, 80, 9]
listaMaiores = []

for i in range(3):
    maior = max(lista)
    listaMaiores.append(maior)
    lista.remove(maior)

print(listaMaiores)
```

**Q06** - Faça um programa que, dadas duas listas de mesmo tamanho, crie uma nova lista com cada elemento igual a soma dos elementos da lista 1 com os da lista 2, na mesma posição.

Exemplo: dadas lista1 = [1, 4, 5] e lista2 = [2, 2, 3], então lista3 = [1+2, 4+2, 5+3] = [3, 6, 8]

```python
lista1 = [1, 4, 5]
lista2 = [2, 2, 3]
lista3 = []

for idx, item in enumerate(lista1):
    soma = lista1[idx] + lista2[idx]
    lista3.append(soma)
print(lista3)
```

**Q07** - Faça um programa que dadas duas listas de mesmo tamanho, imprima o produto escalar entre elas.

OBS: produto escalar é a soma do resultado da multiplicação entre o número na posição i da lista1 pelo número na posição i da lista2, com i variando de 0 ao tamanho da lista.

```python
lista1 = [1, 4, 5]
lista2 = [2, 2, 3]
lista3 = []

for idx, item in enumerate(lista1):
    soma = lista1[idx] * lista2[idx]
    lista3.append(soma)
print(lista3)
```

**Q12** - Faça um programa que sorteia 10 números entre 0 e 100 e conte quantos números sorteados são maiores que 50.

```python
import random

numSorteados = []
cont = 0

for num in range(10):
    sorteado = random.randint(0, 100)
    numSorteados.append(sorteado)

for i in numSorteados:
    if i > 50:
        cont += 1

print(cont)
```

**Q13** - Desafio 1 | Faça um programa que peça para o usuário digitar o nome e a idade de um aluno e o número de provas que esse aluno fez. Depois, o programa deve pedir para o usuário digitar as notas de cada prova do aluno. Ao final o programa deve imprimir uma lista contendo:

a. Nome do aluno na posição 0

b. Idade do aluno na posição 1

c. Uma lista com todas as notas na posição 2

d. A média do aluno na posição 3

e. True ou False, caso a média seja maior que 5 ou não, na posição 4

Dica: Use o que você fez nos exercícios anteriores para criar esse programa.

```python
nome = input('Nome do aluno: ')
idade = int(input('Idade do aluno: '))
provas = int(input('Quantas provas: '))
notas = []
info_aluno = []
cont = 1

while cont <= provas:
    notas.append(float(input(f'Qual a nota da prova {cont}: ')))
    cont += 1

media = sum(notas) / provas

info_aluno.append(nome)
info_aluno.append(idade)
info_aluno.append(notas)
info_aluno.append(media)
info_aluno.append(True if media>=5 else False)

print(info_aluno)
```

**Q14** - Faça um programa como o do item anterior, porém que imprima a média sem considerar a maior e menor nota do aluno (nesse caso o número de provas precisa ser obrigatoriamente maior que dois).

Dica: crie uma cópia com a lista de todas as notas antes de fazer a média.

```python
nome = input('Nome do aluno: ')
idade = int(input('Idade do aluno: '))
provas = int(input('Quantas provas: '))
notas = []
info_aluno = []
cont = 1

while provas <= 2:
    provas = int(input('Quantas provas: '))

while cont <= provas:
    notas.append(float(input(f'Qual a nota da prova {cont}: ')))
    cont += 1

copia_notas = notas.copy()
copia_notas.remove(min(copia_notas))
copia_notas.remove(max(copia_notas))

# print(notas)
# print(copia_notas)
media = sum(copia_notas) / len(copia_notas)

info_aluno.append(nome)
info_aluno.append(idade)
info_aluno.append(notas)
info_aluno.append(round(media,2))
info_aluno.append(True if media>=5 else False)

print(info_aluno)
```
