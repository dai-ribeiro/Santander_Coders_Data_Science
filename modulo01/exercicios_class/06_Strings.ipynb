Q03 - Altere o exercício anterior para que a string copiada alterne entre letras maiúsculas e minúsculas.

Exemplo: se o usuário digitar "latex" o programa deve imprimir "LaTeX".

```python
palavra = input('Digite uma palavra: ')
nova_palavra = ''

for idx, letra in enumerate(palavra):
    if idx % 2 == 0:
        nova_palavra += letra.upper()
    else:
        nova_palavra += letra.lower()
print(nova_palavra)
```

Q05 - Faça uma função que receba uma string e retorne uma nova string substituindo:

'a' por '4'

'e' por '3'

'I' por '1'

't' por '7'

```python
def trocaLetra(palavra):
    nova_palavra = ''
    for letra in palavra:
        if letra == 'a' or letra == 'A':
            nova_palavra += '4'
        elif letra == 'e' or letra == 'E':
            nova_palavra += '3'
        elif letra == 'i' or letra == 'I':
            nova_palavra += '1'
        elif letra == 't' or letra == 'T':
            nova_palavra += '7'
        else:
            nova_palavra += letra
    return nova_palavra


palavra = input('Digite uma palavra: ')
trocaLetra(palavra)

```

Q09 - Faça uma função que receba uma string que contém tanto números quanto letras e caracteres especiais, e que separe as letras em uma variável e os números em outra (os caracteres especiais podem ser descartados). Ao final a função deve imprimir as duas variáveis.

```python
def separaNumLetra(string):
    letra = ''
    num = ''
    for caracter in string:
        if caracter.isnumeric():
            num += caracter
        elif caracter.isalpha():
            letra += caracter
    return f'Letras: {letra}; Números: {num}'

separaNumLetra('a1b2b3b4&f9*')
```

Q10 - Desafio - Faça uma função que receba uma string e uma letra e:

a. imprima quantas vezes a letra aparece na string;

b. imprima todas as posições em que a letra aparece na string;

c. retorne a distância entre a primeira e a última aparição dessa letra na string.

```python
## A
def contaLetra(string, letra):
    cont = 0
    for caracter in string:
        if caracter == letra:
            cont += 1
    return f'A letra {letra} aparece {cont} vez(es).'

print(contaLetra('banana', 'a'))
```

```python
## B
def posicaoLetra(string, letra):
    posicao = []
    string = list(string)
    for idx, caracter in enumerate(string):
        if caracter == letra:
            posicao.append(idx)
    if posicao == []:
        return f'A letra não aparece na palavra.'
    else:
        return f'A letra {letra} aparece nas posições {posicao}.'


print(posicaoLetra('banana', 'a'))
```

```python
## C

def distanciaLetra(string, letra):
    posicao = []
    string = list(string)
    distancia = 0
    for idx, caracter in enumerate(string):
        if caracter == letra:
            posicao.append(idx)
    distancia = posicao[-1] - posicao[0]
    return distancia

print(distanciaLetra('banana', 'a'))
```
