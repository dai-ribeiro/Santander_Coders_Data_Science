**Q01** - Faça um programa que peça ao usuário um número e imprima todos os números de um até o número dado.

Exemplo:

digite: 5

imprime: 1 2 3 4 5

```python
num = int(input('Informe um número: '))
cont = 1

while cont <= num:
  print(cont, end=' ')
  cont += 1
```

**Q02** - Peça ao usuário para digitar um número e imprima o fatorial de n.

```python
num = int(input('De qual número deseja saber o fatorial? '))
fat = 1
aux = 1

while aux <= num:
  fat *= aux
  aux += 1

print(f'O fatorial de {num} é: {fat}')
```

**Q03** - Peça ao usuário para digitar um número N e some todos os números de 1 a N utilizando o laço de repetição while.

```python
num = int(input('Informe um número: '))
soma = 0
cont = 1

while cont <= num:
  soma += cont
  cont += 1

print(soma)
```

**Q04** - Faça um programa que imprima a tabuada do 9 (de 9x1 a 9x10) usando loops.

```python
cont = 1

while cont <= 10:
  valor = 9 * cont
  print(f'9 x {cont:2d} = {valor:2d}')
  cont += 1
```

**Q05** - Faça um programa que recebe um número inteiro do usuário e imprime na tela a quantidade de divisores desse número e quais são eles.

Exemplo 1
input: 10
output:
Divisor: 1
Divisor: 2
Divisor: 5
Total de divisores: 3

Exemplo 2
input: 15
output:
Divisor: 1
Divisor: 3
Divisor: 5
Total de divisores: 3

```python
num = int(input('Informe um número: '))
div = 1
cont = 0

while div <= num:
  if (num % div) == 0:
    print(f'Divisor: {div}')
    cont += 1
  div += 1

print(f'Total de divisores: {cont}')
```

**Q06** - Faça um programa, usando loops, que peça para um usuário digitar um número e que só finaliza quando o usuário digitar 0. Ao final imprima a soma de todos os números digitados.

```python
num = int(input('Digite um número: '))
soma = num

while num != 0:
  num = int(input('Digite um número: '))
  soma += num
print(f'A soma de todos os números é {soma}')
```

**Q07** - Faça um programa que sorteia um número N e peça para o usuário adivinhar o número sorteado. A cada resposta errada, o seu programa deve imprimir um aviso dizendo que a resposta está errada e pedir novamente uma resposta ao usuário.

Para a realização desse exercício, utilize alguma função da biblioteca random (e.g. randint()).

```python
import random

numSorteado = random.randint(1,50)

numInput = int(input('Qual o número sorteado (de 1 a 50): '))

while numSorteado != numInput:
  numInput = int(input('Você errou! Tente novamente: '))
else:
  print(f'Você acertou, o número sorteado foi {numSorteado}.')
```

**Q08** - Faça um programa que peça para o usuário digitar a idade, o salário e o sexo de uma pessoa até que as entradas digitadas sejam válidas.

a. Idade: entre 0 e 150;

b. Salário: maior que 0;

c. Sexo: M, F ou Outro.

```python
idade = input('Qual a sua idade? ')

while not idade.isdigit():
  idade = input('Digite apenas números. \nQual a sua idade? ')
else:
  idade = int(idade)
  while (idade < 0) or (idade > 150):
    idade = int(input('Idade inválida! Qual a sua idade (de 0 a 150)?'))

salario = input('Qual o seu salário? ')

while not salario.isdigit():
  salario = input('Digite uma valor válido. \nQual o seu salário? ')
else:
  salario = int(salario)
  while salario <= 0:
    salario = float(input('Salário Inválido! Qual o seu salário? '))

sexo = input('Qual o seu sexo? (M, F ou OUTRO) ')
sexo = sexo.upper()

while (sexo != 'M') and (sexo != 'F') and (sexo != 'OUTRO'):
  sexo = input('Sexo inválido! Qual o seu sexo? ').upper()

print(f'Idade inserida: {idade}')
print(f'Salário inserido: {salario}')
print(f'Sexo inserido: {sexo}')
```

**Q09** - Desafio! - Calcule a soma de até mil termos da série 1 + 1/2 + 1/4 + 1/8 + 1/16 + ...

Dica: Use três variáveis:

• um contador, que começa em zero;

• uma variável para a soma de todos os termos, que também começa em zero;

• uma variável para cada termo, que começa em 1 e a cada loop é dividida por 2.

A repetição da soma de mil termos pode ser feita com a função while contador < 1000.

```python
soma = 0
aux = 1
cont = 0

while cont < 1000:
  soma += aux
  aux /= 2
  cont += 1

print(soma)
```

**Q10** - Super Desafio! - Calcule a soma de mil termos dos inversos dos fatoriais: 1/(1!) + 1/(2!) + 1/(3!) + 1/(4!) + ...

Dica: Assim como no exercício anterior use três variáveis: um contador; uma variável para a soma; e uma variável para os termos. Lembre-se de que 4! = 4*3*2*1 que também é igual a 4*3!.

```python
soma = 0
fatorial = 1
cont = 1

while cont <= 1000:
  fatorial *= cont
  soma += (1 / fatorial)
  cont += 1

print(soma)
```
